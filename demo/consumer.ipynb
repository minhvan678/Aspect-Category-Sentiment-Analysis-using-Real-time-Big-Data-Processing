```python
from kafka import KafkaConsumer
from pymongo import MongoClient
from json import loads
```

```python
topic_name = 'traveloka'
kafka_server = 'localhost:9092'
```

```python
consumer = KafkaConsumer(
    topic_name,
    bootstrap_servers=kafka_server,
    auto_offset_reset = 'latest',
    enable_auto_commit = True,
    group_id = 'my-group', 
    value_deserializer = lambda x: loads(x)
)
```

```python
client = MongoClient("mongodb+srv://nhq188:fiora123@cluster0.vcb5k8u.mongodb.net/")
collection = client.traveloka.comment
```

```python
import time

for message in consumer:
    try:
        mess = loads(message.value)
        if mess['comment'] == None:
            mess['comment'] = ""
        collection.insert_one(mess)
        print(f"{mess} added to {collection}")
        time.sleep(1)
    except:
        mess = message.value
        if mess['comment'] == None:
            mess['comment'] = ""
        collection.insert_one(mess)
        print(f"{mess} added to {collection}")
        time.sleep(1)
    else:
        pass
```
